## Description of Dataset

```python
raw_df.info()
```

Column description:



*   **age**: patients age in years
*   **sex**:  Gender (Male : 1; Female : 0)
*   **cp**: Type of chest pain experienced by patient. This term categorized into 4 category. 0: typical angina, 1: atypical angina, 2: non- anginal pain, 3: asymptomatic
*   **trestbps**:  patient's level of blood pressure at resting mode in mm/HG
*   **chol**: Serum cholesterol measurement in mg/dl
*   **fbs**:  Fasting Blood sugar levels. > 120 mg/dl: 1. < 120 mg/dl: 0
*   **restecg**:  Result of electrocardiogram while at rest are represented in 3 distinct values
0 : Normal. 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV) 2: showing probable or definite left ventricular hypertrophyby Estes' criteria.
*   **thalach**: Maximum heart rate achieved
*   **exang**: angina induced by Exercise.  (1 = yes; 0 = no)
*   **oldpeak**: Exercise induced ST-depression in relative with the state of rest.
*   **slope**: slope of the peak exercise ST segment (0: up sloping; 1: flat; 2: down sloping)
*   **ca**: number of major vessels colored by flourosopy (0–3)
*   **thal**: A blood disorder called thalassemia
1: normal blood flow 2: fixed defect (no blood flow in some part of the heart) 3: reversible defect (a blood flow is observed but it is not normal)
*   **target**: 1 = patient is suffering from heart disease and 0 = patient is normal

# Building Models

```python
from treeviz import tree_print
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

## Splitting Dataset

```python
X = df.drop(['target'],axis=1)
Y = df['target']

datasets = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=2)
(X_train, X_test, Y_train, Y_test) = datasets
```

## Low Complexity Model

```python
model = tree.DecisionTreeClassifier(criterion='entropy', max_depth=1, random_state=1)

model.fit(X_train, Y_train)
tree_print(model,X)

Y_train_model = model.predict(X_train)
print("Train Accuracy: {:3.2f}".format(accuracy_score(Y_train, Y_train_model)))

Y_test_model = model.predict(X_test)
print("Test Accuracy: {:3.2f}".format(accuracy_score(Y_test, Y_test_model)))
```

The low complexity model is underfitting because the decision tree's depth of 1 means it can only make one decision based on only one feature, 'thal'. While simplicity can be an advantage, in this case, the model is overly simplistic for the complexity of the problem. Heart disease diagnosis typically involves multiple features and complex interactions.

Also, the model is biased because it simplifies the problem too much. It essentially assumes that 'thal' alone can provide a diagnosis of heart disease, which is unrealistic. Additionally, it doesn't take into account the complexity of real-world medical data, where multiple factors are typically involved in making a diagnosis.

<br>

The model only makes one comparison: if the patient has normal blood flow (score of 1), they are normal, but if the patient has a fixed defect (score of 2) or reversible defect (score of 3), they have a heart disease.

The only decision that the model can make is whether someone has a heart disease or is normal based on the presence or absence of a blood disorder.

## Unrestricted High Complexity Model

```python
model = tree.DecisionTreeClassifier(criterion='entropy', random_state=1)

model.fit(X_train, Y_train)
tree_print(model,X)

Y_train_model = model.predict(X_train)
print("Train Accuracy: {:3.2f}".format(accuracy_score(Y_train, Y_train_model)))

Y_test_model = model.predict(X_test)
print("Test Accuracy: {:3.2f}".format(accuracy_score(Y_test, Y_test_model)))
```

## Graphically Comparing Train and Test Scores

```python
data = {'Complexity': ['Low','High'], 'Training Accuracy': [0.77, 1.00], 'Test Accuracy': [0.74, 0.75]}

model_df = pd.DataFrame(data)

plt.plot(model_df['Complexity'], model_df['Training Accuracy'], marker='o', label='Training Accuracy', linestyle='-', color='blue')
plt.plot(model_df['Complexity'], model_df['Test Accuracy'], marker='o', label='Test Accuracy', linestyle='-', color='orange')

plt.xlabel('Model Complexity')
plt.ylabel('Accuracy')
plt.title('Training and Test Accuracy vs. Complexity')

plt.legend()
plt.show()
```

The graph above provides clear evidence of overfitting in the high complexity model. This is evident as the training accuracy reaches a perfect 100%, while the testing accuracy exhibits only marginal improvement. A training score of 100% strongly indicates overfitting because it implies that the model is reliant on specific details within the training dataset, making it less effective when making predictions on the test dataset by diminishing its ability to generalize effectively to new, unseen data points.

To prevent overfitting, I will add a constraint by using the max_depth parameter and set it at a value that does not underfit nor overfit the model. By carefully selecting an appropriate value for max_depth, I can find a balance between model underfitting and overfitting.

# Model Search

## Grid Search with Cross-Validation

```python
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score

cv = KFold(n_splits=5,  shuffle = True)


model = tree.DecisionTreeClassifier()
param_grid = {
    'max_depth': list(range(1,11)),
    'criterion': ['entropy', 'gini']
    }
grid = GridSearchCV(model, param_grid, cv=5)

grid.fit(X_train, Y_train)

print("Best parameters: {}".format(grid.best_params_))
```

According to the grid search with cross validation, the best set of hyperparameters for the model is a criterion of entropy and a max_depth of 4.
Best parameters: {'criterion': 'entropy', 'max_depth': 4}

<br>

I used cross-validation to find the best hyperparameter values for my model. It systematically explores different combinations of hyperparameters and identifies the ones that lead to optimal performance on unseen data. Cross-validation provided me with a more reliable estimate of how well my model is likely to perform on data it has never seen before. This is crucial to ensure that my model generalizes well and doesn't overfit to the training data.







## Cross Validation Scores

```python
dtree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4)

scores = cross_val_score(dtree, X_train, Y_train,cv=cv)
print("Fold Accuracies: {}".format(scores))
print("Accuracy: {:3.2f}".format(scores.mean()))
```

The cross validation scores of the best model are: 81%, 88%, 76%, 67%, and 76%.

The mean accuracy of the best model is 78%.

The cross validation scores have an average training accuracy of 78%, which is in the between of the low and high complexity model. Based on this observation, I believe the medium complexity model is the best model. It has a balance with its training accuracy, it's not excessively high to indicate overfitting, nor is it too low to suggest underfitting. This middle-ground accuracy level indicates that the medium complexity model has the potential to generalize well to unseen data, making it an effective choice to predict whether a person has a heart disease or not.

## Comparing Both Models

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,balanced_accuracy_score, confusion_matrix

def model_metrics(y_true, y_predict):
  accuracy= accuracy_score(y_true, y_predict)
  precision=precision_score(y_true, y_predict, average='micro')
  recall=recall_score(y_true, y_predict, average='micro')
  f1=f1_score(y_true, y_predict, average='micro')
  balanced_accuracy=balanced_accuracy_score(y_true, y_predict)
  return [accuracy, precision, recall, f1, balanced_accuracy]
```

```python
medium_train_metrics= model_metrics(Y_train, pred_train_medium)
medium_test_metrics = model_metrics(Y_test, pred_test_medium)
```

```python
ten_percent_train_metrics= model_metrics(Y_train, pred_train_ten_percent)
ten_percent_test_metrics = model_metrics(Y_test, pred_test_ten_percent)
```

```python
df_medium = pd.DataFrame([ medium_train_metrics, medium_test_metrics],  columns=['Accuracy', 'Pecision', 'Recall', 'F1 score', 'Balanced accuracy'],
                      index =['Train', 'Test'] )
df_ten_percent = pd.DataFrame([ ten_percent_train_metrics, ten_percent_test_metrics],  columns=['Accuracy', 'Pecision', 'Recall', 'F1 score', 'Balanced accuracy'],
                      index =['Train', 'Test'] )
all_metrics_values={"Medium Complexity": df_medium,
    "Ten Percent Worse": df_ten_percent}
all_metrics_values
```

```python
import plotly.graph_objects as go

def plot_bar_results(dataset_type, all_metrics_values):
    fig = go.Figure()
    for model_name, metrics_df in all_metrics_values.items():
        metrics_names = list(metrics_df.columns)
        values = metrics_df.loc[dataset_type].to_numpy()
        fig.add_trace(go.Bar(y=values, x=metrics_names, name=model_name))
    fig.update_layout(barmode="group", xaxis_tickangle=-45, title_text=dataset_type+" Dataset")
    fig.show()
```

```python
plot_bar_results("Train", all_metrics_values)
```

```python
plot_bar_results("Test", all_metrics_values)
```

These bar plots show that both models have a high scores, which indicates that these models would be good at predicting whether someone has a heart disease or not based on their measurments on the features in this dataset.

**Recall, Precision, and F1 score**: Both of these scores are very high. ## Confidence Interval

```python
def classification_confint(acc, n):
    '''
    Compute the 95% confidence interval for a classification problem.
      acc -- classification accuracy
      n   -- number of observations used to compute the accuracy
    Returns a tuple (lb,ub)
    '''
    import math
    interval = 1.96*math.sqrt(acc*(1-acc)/n)
    lb = max(0, acc - interval)
    ub = min(1.0, acc + interval)
    return (lb,ub)

def regression_confint(rs_score, n, k):
    '''
    Compute the 95% confidence interval for a regression problem.
      rs_score -- R^2 score
      n        -- number of observations used to compute the R^2 score
      k        -- number of independent variables in dataset
    Returns a tuple (lb,ub)
    Reference:
    https://books.google.com/books?id=gkalyqTMXNEC&pg=PA88#v=onepage&q&f=false
    '''
    import math
    interval = 2*math.sqrt((4*rs_score*(1-rs_score)**2*(n-k-1)**2)/((n**2 - 1)*(n+3)))
    lb = max(0, rs_score - interval)
    ub = min(1.0, rs_score + interval)
    return (lb,ub)
```

```python
observations = len(X_test)

# first classifier
acc1 = .77
lb1,ub1 = classification_confint(acc1,observations)
print('Medium Complexity Accuracy: {} ({:3.2f},{:3.2f})'.format(acc1,lb1, ub1))

# second classifier
acc2 = .78
lb2,ub2 = classification_confint(acc2,observations)
print('Ten Percent Worse Accuracy: {} ({:3.2f},{:3.2f})'.format(acc2,lb2, ub2))
```
